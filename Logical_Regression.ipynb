##1. Introduction to the Example

What is the Problem?  
We want to build a machine learning model to classify emails as spam or not spam, a classic binary classification task.  

Target Variable (y): Categorical, with two classes: “spam” or “not spam.”  

Features (X): Numeric representations of email characteristics, like the frequency of certain words (e.g., “win,” “free”) or email length.  

Real-World Relevance: Spam filters save time and protect users, used by companies like Gmail and Yahoo.

Why Classification?  
As discussed, classification predicts categorical targets (unlike regression’s continuous targets, e.g., house prices).  

This example shows how a model learns patterns from labeled emails to predict new ones, linking to our prior discussion of classification’s categorical target variable.

Analogy: It’s like training a librarian to sort books into “fiction” or “non-fiction” by studying examples of each type. The librarian learns patterns to classify new books.
Tool: We’ll use Scikit-learn, a beginner-friendly Python library, to build a logistic regression model for classification (not to be confused with linear regression for regression tasks).



##2. Python Code Explanation and Demo

Setup and Context:  
We’ll use a simplified dataset of emails with features (e.g., word frequencies) and labels (spam = 1, not spam = 0).  

For teaching, we’ll create a small synthetic dataset to keep it manageable, but in real projects, you’d use a larger dataset (e.g., from Kaggle).  

Prerequisites: Students should have basic Python knowledge (lists, arrays) and Scikit-learn installed (pip install scikit-learn).



```python
# 1. Import libraries
from sklearn.linear_model import LogisticRegression
import numpy as np

# 2. Create Features (X) and Labels (y)
X = np.array([
    [2, 3, 50],  # Spam (Many "win" and "free" words, longer email)
    [0, 1, 30],  # Not Spam
    [1, 2, 60],  # Spam
    [0, 0, 25],  # Not Spam
    [3, 4, 70]   # Spam
])

y = np.array([1, 0, 1, 0, 1])  # 1=Spam, 0=Not Spam

# 3. Build the Logistic Regression model
spam_detector = LogisticRegression()
spam_detector.fit(X, y)

# 4. Predict a new email
new_email_features = np.array([[2, 2, 55]])  # (2 wins, 2 frees, 55 words)
prediction = spam_detector.predict(new_email_features)
prediction_prob = spam_detector.predict_proba(new_email_features)

# 5. Display results
if prediction[0] == 1:
    print("The new email is predicted as: SPAM")
else:
    print("The new email is predicted as: NOT SPAM")

print(f"Probability - Spam: {prediction_prob[0][1]*100:.1f}%, Not Spam: {prediction_prob[0][0]*100:.1f}%")
```

Libraries:  
numpy: Handles arrays for data.  

sklearn.linear_model.LogisticRegression: Builds the classification model.

Data:  
X: 5 emails, each with 3 features (frequency of “win,” “free,” email length).  

y: Labels (1 for spam, 0 for not spam).

Model:  
LogisticRegression(): Creates a model that learns to separate spam from not spam.  

model.fit(X, y): Trains the model to find patterns (e.g., high “free” frequency → spam).

Prediction:  
new_email: A new email with features [2, 2, 55].  

model.predict(): Outputs the predicted class (1 or 0).  

model.predict_proba(): Shows the probability of each class (e.g., 80% spam, 20% not spam).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit  # Sigmoid function

# 1. Create a range of input values
X = np.linspace(-10, 10, 100)

# 2. Calculate the sigmoid (S-curve)
y = expit(X)  # Logistic (sigmoid) function

# 3. Plot
plt.figure(figsize=(8, 5))
plt.plot(X, y, label='Sigmoid Curve', color='blue')

# 4. Add threshold line at 0.5
plt.axhline(0.5, color='red', linestyle='--', label='Decision Threshold (0.5)')

# 5. Labels and title
plt.title('Logistic Regression: Sigmoid Curve and Decision Threshold')
plt.xlabel('Feature Combination (weighted sum)')
plt.ylabel('Probability of Spam')
plt.legend()
plt.grid(True)

# 6. Show plot
plt.show()
```
